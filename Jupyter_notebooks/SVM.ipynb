# Máquinas de vector soporte (SVM)

Son algoritmos que su entrada son datos y su salida son líneas que separan los datos.

Ahora bien, hay líneas que separan datos mejor que otras, y las que mejor separan los datos son aquellas donde se maximiza la distancia al punto más cercano EN TODAS AMBAS  DE DATOS, y esto se conoce como margen.
Entre mayor sea la margen --> se maximiza la ROBUSTEZ del algoritmo.

**Las SVM ponen como primer objetivo la correcta clasificación de los datos y como segundo objetivo la maximización de la margen**
<img src="svm.png">
Las ventajas de una SVM:

Eficaz en espacios de alta dimensión.

Eficaz en casos en que el número de dimensiones es mayor que el número de muestras.

Es eficiente en la memoria.

Una de las desventajas es :

Si el número de características (features) es mucho mayor que el número de muestras, es probable que el método dé un rendimiento pobre.

Para usar este algoritmo (y otros) debemos de seguir siempre 4 pasos:

1. importar el algoritmo
2. Crear el clasificador
3. Entrenar el clasificador
4. Hacer predicciones

```python
#1 Importamos el algoritmo SVM y unos sets de datos para trabajar con ellos
from sklearn import svm, datasets
import numpy as np
import matplotlib.pyplot as plt
#vamos a usar los datos de unas plantas
iris = datasets.load_iris()
# apenas vamos a usar 2 caracteristicas o labels de estas plantas (ancho y largo de la hoja)
X = iris.data[:, :2]
# tenemos 3 tipos de planta rotulada (labels): 0, 1 y 2
y = iris.target
```

```python
#2 creamos el clasificador
clf = svm.SVC(kernel='linear')
```

```python
#3 entrenamos el algoritmo, con fit
clf.fit(X, y)
```

```python
# no vamos a predecir nada en este ejemplo pero vamos a visualizar la información

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel('Largo de hoja')
plt.ylabel('Ancho de hoja')
plt.title('SVM con Kernel lineal')

plt.show()
```
